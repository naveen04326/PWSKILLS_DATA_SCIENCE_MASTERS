```python
# Ans 7 : 
text_emldata = {"names" : ["maddy@gmail.com","Madhavan@gmail.com","madhavachari@gmail.com"]}

df_emltext = pd.DataFrame(text_emldata)

print(df_emltext,"\n")

def user_name(x):
    y=x.split("@")
    return(y[0])

df_emltext["user_name"]= df_emltext["names"].apply(user_name)

print("\n", df_emltext,"\n")
```
