```python
# Ans 1 : 

from functools import reduce




def list_conv(list1) :
    
    list_len = len(list1)

    print("No of elements in the list are :  ", list_len)

    new_list = []

    for i in range(list_len) :
        if type(list1[i]) == int :
            new_list.append(list1[i])
        
        elif type(list1[i]) == set :
        
            set_list = list(list1[i])
            for s in range(len(set_list)):
                if type(set_list[s]) == int :
                    new_list.append(set_list[s])
               
                elif type(set_list[s]) == tuple :
                    stup_list = list(list1[i])
                    for t in range(len(stup_list)):
                        if type(stup_list[s][t]) == int :
                            new_list.append(stup_list[s][t])
          
        elif type(list1[i]) == dict :
       
            dict_item={}
            dict_item= list1[i]
         
            keys_list= list (dict_item.keys())
            #iterate throuh values to breal tuples and list
            values_list = list(dict_item.values())
        
            dict_list =keys_list+values_list
        
          
            for d in range(len(dict_list)):
                if type(dict_list[d]) == int :
                    new_list.append(dict_list[d])
        
                elif type(dict_list[d]) == tuple:
                    dtup_list = list(dict_list[d])
                    for t in range(len(dtup_list)):
                        if type(dtup_list[t]) == int :
                            new_list.append(dtup_list[t])
                
                elif type(dict_list[d]) == list:
                     for l in range(len(dict_list[d])):
                        if type(dict_list[d][l]) == int :
                            new_list.append(dict_list[d][l])

        
        elif type(list1[i]) == tuple:
            tup_list = list(list1[i])
            for t in range(len(tup_list)):
                if type(tup_list[t]) == int :
                    new_list.append(tup_list[t])
                
        elif type(list1[i]) == list:
             for l in range(len(list1[i])):
                if type(list1[i][l]) == int :
                    new_list.append(list1[i][l])
        else :
            continue
        
    return new_list


test_list = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45,22, 61, 34)}, [56, 'data science'], 'Machine Learning']

output_list = list_conv(test_list)
product = reduce(lambda x,y : x*y, output_list)
print ("The list having all the intigers is  : ", output_list)
print("The product of all the intigers in the list is : ", product)
```

```python
#Ans 2 : Encrypting Msg : 

def encrypt_Msg(message):
    
    message = str(message).lower()
    encrypt_msg=''
    for i in message :
       
        temp= ord(i)
        if (temp >= 97 and int(ord(i)) <= 122):
            encrypt_msg  += chr(122+97-ord(i))
                       
        elif( i == " "):
            encrypt_msg  += "$"
            
        else:
            encrypt_msg  += i              
        
    
    return encrypt_msg.lower()


input_Sentence= 'I want to become a Data Scientist.'
print("original message = ", input_Sentence)
encrypt_msg = encrypt_Msg(input_Sentence)
print("\nencripted message = ", encrypt_msg)
    
```
