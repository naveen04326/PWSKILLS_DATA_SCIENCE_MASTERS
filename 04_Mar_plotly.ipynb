```python
# Ans 1 : 

import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px

df = sns.load_dataset('titanic')
fig = px.scatter(df, x="age", y="fare")
fig.show()
```

```python
# Ans 2 : 

import plotly.express as px
import plotly as plt

df1 = plt.data.tips()
print(df1.head())
fig = px.box(df1, y= 'total_bill')
fig.show()
```

```python
# Ans 3 : 
import plotly.express as px
import plotly as plt

df1 = plt.data.tips()
print(df1.head())
fig = px.histogram(df1, x = 'sex', y= 'total_bill',pattern_shape='smoker', color= 'day')
fig.show()
```

```python
# Ans 4 : 
import plotly.express as px
import plotly as plt

df2 = plt.data.iris()
print(df2.head())

fig = px.scatter_matrix(df2, dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                        color = 'species',height = 500)
fig.show()
```

```python
# Ans 5 : 
"""
Distplots
The distplot figure factory displays a combination of statistical representations of numerical data,
such as histogram, kernel density estimation or normal curve, and rug plot.

The distplot can be composed of all or any combination of the following 3 components −

a. histogram
b. curve: (a) kernel density estimation or (b) normal curve, and
c. rug plot
The figure_factory module has create_distplot() function which needs a mandatory parameter called hist_data.

"""
from plotly import figure_factory  as  ff
import numpy as np

x = np.random.randn(1000)
hist_data = [x]
group_labels = ['distplot']
fig = ff.create_distplot(hist_data, group_labels)
fig.show()
```
